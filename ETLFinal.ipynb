```python
file_to_load="all_stocks_5yr.csv"
```

```python
stock_csv_data=pd.read_csv(file_to_load)
stock_csv_data.head()
```

```python
stock_csv_df=pd.DataFrame(stock_csv_data[["Name","date","open","close","high","low"]])
stock_csv_df=stock_csv_df.rename(columns={"Name":"NAME","date":"DATE","open":"OPEN","close":"CLOSE","high":"HIGH","low":"LOW"})
stock_csv_df.head()
```

```python
stock_csv_df=stock_csv_df[stock_csv_df["NAME"].isin(["AAL"])]
stock_csv_df.sort_values(by='DATE',ascending=False)
#df.sort_values(by='col1', ascending=False)

stock_csv_df.head()
```

```python
#getting data from API
stock_json_data=[]

change=json['FB']['chart']
news=json['FB']['quote']['symbol']

for i in range(0,len(change)):
    change_data=change[i]["change"]
    open_data=change[i]['open']
    volume_data=change[i]['volume']
    close_data=change[i]["close"]
    date_data=change[i]['date']
    high_data=change[i]['high']
    low_data=change[i]['low']
    title=news
    stock_json_data.append([title,date_data,open_data,close_data,high_data,low_data,change_data,volume_data])
#stock_json_df['date']=pd.to_datetime(stock_json_df['date'])
stock_json_data

```

```python
#creating dataframe
stock_json_df=pd.DataFrame(stock_json_data,columns=['Name',"date","open","close","high","low","Change","Volume"])
stock_json_df=stock_json_df.rename(columns={"Name":"NAME","date":"DATE","open":"OPEN","close":"CLOSE","high":"HIGH","low":"LOW","Change":"CHANGE","Volume":"VOLUME"})
stock_json_df.head()
```

```python
# cleaning dataframe
stock_json_df=stock_json_df.drop(['CHANGE','VOLUME'],axis=1)
stock_json_df.head()
```

```python
combined_data=stock_json_df.append(stock_csv_df)
combined_data.head()
```

```python
data_excel=combined_data.to_excel("data_excel.xlsx")
data_excel
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.combined_data
collection = db.combined_data
FinalData=json.loads(combined_data.to_json(orient='records'))
collection.insert_many(FinalData)
cursor = list(collection.find({}))
for document in cursor:
 print(document)
```
